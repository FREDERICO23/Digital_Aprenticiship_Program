```python
# Downloading the dataset 

import requests

download_url ='https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json'
target_json_path = 'water_points.json'

response = requests.get(download_url)
response.raise_for_status() #check if request was successful
with open(target_json_path, 'wb') as f:
    f.write(response.content)

print('Setup complete.')
```

```python
import pandas as pd
resources = pd.read_json('water_points.json')
type(resources)
```

```python
# NUMBER OF WATER POINTS THAT ARE FUNCTUIONAL

is_functioning = resources['water_point_condition']=='functioning'
is_functioning.value_counts()
```

```python
# NUMBER OF WATER POINTS PER COMMUNITY

water_points = resources.groupby('communities_villages').water_point_condition.value_counts()
water_points
```

```python
# COMMUNITY RANK BY BROKEN WATER POINTS in PERCENTAGE
broken_water_points = resources.loc[resources['water_point_condition']=='broken','communities_villages'].value_counts('broken','communities_villages')
broken_water_points

```
